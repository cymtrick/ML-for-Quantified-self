```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
processed_X_train = pd.read_csv('processed_X_train_chapter3.csv')
#len(processed_X_train.columns)
processed_X_test = pd.read_csv('processed_X_test_chapter3.csv')

# performs t-sne with different perplexity values and their repective plots..

def perform_tsne(X_data, y_data, perplexities, n_iter=1000, img_name_prefix='t-sne'):
        
    for index,perplexity in enumerate(perplexities):
        # perform t-sne
        print('\nperforming tsne with perplexity {} and with {} iterations at max'.format(perplexity, n_iter))
        X_reduced = TSNE(verbose=2, perplexity=perplexity).fit_transform(X_data)
        print('Done..')
        
        # prepare the data for seaborn         
        print('Creating plot for this t-sne visualization..')
        df = pd.DataFrame({'x':X_reduced[:,0], 'y':X_reduced[:,1] ,'label':y_data})
        
        # draw the plot in appropriate place in the grid
        plt.figure(figsize=(16,10))
        sns.scatterplot(
            x="x", y="y",
            hue="label",
            palette=sns.color_palette("hls", 12),
            data=df,
            legend="full"
        )
        img_name = img_name_prefix + '_perp_{}_iter_{}.png'.format(perplexity, n_iter)
        print('saving this plot as image in present working directory...')
        plt.savefig(img_name)
        plt.show()
        print('Done')
```

```python
#knn
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn import preprocessing
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split

def loadData(filename):
    X = np.loadtxt(filename)
    return X
def total_count(y):
    count = np.zeros(12)
    for i in range(y.size):
        if (y[i] == 1):
            count[0] += 1
        if (y[i] == 2):
            count[1] += 1
        if (y[i] == 3):
            count[2] += 1
        if (y[i] == 4):
            count[3] += 1
        if (y[i] == 5):
            count[4] += 1
        if (y[i] == 6):
            count[5] += 1
        if (y[i] == 7):
            count[6] += 1
        if (y[i] == 8):
            count[7] += 1
        if (y[i] == 9):
            count[8] += 1
        if (y[i] == 10):
            count[9] += 1
        if (y[i] == 11):
            count[10] += 1
        if (y[i] == 12):
            count[11] += 1
    return count

y_test_count = total_count(y_test)

def confusion_matrix(y_test, y_pred):
    cm = np.zeros((12, 12))
    for i in range(12):
        for j in range(12):
            for k in range(y_pred.size):
                if (y_pred[k] == j + 1 and y_test[k] == i + 1):
                    cm[i][j] += 1
    return cm

def pre_rec(cm, count):
    perc = []
    for i in range(12):
        perc.append(float("{0:.2f}".format(cm[i][i] / count[i] * 100)))
    return perc
        
def overall_accuracy(cm, y_test):
    sum = 0
    for i in range(12):
        sum += cm[i][i]
    return float("{0:.2f}".format(sum * 100.0 / y_test.size)) 

def plot_confusion_matrix(cm,lables,img_name):
    fig, ax = plt.subplots(figsize=(12,8)) # for plotting confusion matrix as image
    im = ax.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    ax.figure.colorbar(im, ax=ax)
    ax.set(xticks=np.arange(cm.shape[1]),
    yticks=np.arange(cm.shape[0]),
    xticklabels=lables, yticklabels=lables,
    ylabel='True label',
    xlabel='Predicted label')
    plt.xticks(rotation = 90)
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, int(cm[i, j]),ha="center", va="center",color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
    plt.savefig('figures/'+img_name)
    
def get_best_randomsearch_results(model):
    print("Best estimator : ", model.best_estimator_)
    print("Best set of parameters : ", model.best_params_)
    print("Best score : ", model.best_score_)
    
processed_X_train = pd.read_csv('processed_X_train.csv')
#len(processed_X_train.columns)
processed_X_test = pd.read_csv('processed_x_test.csv')

y_train = loadData('y_train.txt')
y_test = loadData('y_test.txt')

X_train = processed_X_train.drop(['subject', 'Activity','ActivityName','Unnamed: 0'], axis=1)
X_test = processed_X_test.drop(['subject', 'Activity','ActivityName','Unnamed: 0'], axis=1)
#y_pre_knn = le.fit_transform(y_pre_knn)

knc = KNeighborsClassifier(16)
knc.fit(X_train, y_train)
y_predict_knc = knc.predict(X_test)
conf = confusion_matrix(y_test,y_predict_knc )
y_pred_count_knc = total_count(y_predict_knc)
cmatrix_knc = confusion_matrix(y_test, y_predict_knc)
plot_confusion_matrix(conf,labels,'knn_cm')
print("\nK-Neighbors Classifier:")
print(cmatrix_knc)
print("")
print("Best estimator : ", knc.best_estimator_)
print("Best set of parameters : ", knc.best_params_)
recall_knc = pre_rec(cmatrix_knc, y_test_count)
precision_knc = pre_rec(cmatrix_knc, y_pred_count_knc)
accuracy_knc = overall_accuracy(cmatrix_knc, y_test)

print("Precision for KNN: ")
print(precision_knc)
print("Recall for KNN: ")
print(recall_knc)
print("Accuracy for KNN: ")
print(accuracy_knc)

labels = ['WALKING','WALKING_UPSTAIRS','WALKING_DOWNSTAIRS','SITTING','STANDING','LAYING','STAND_TO_SIT','SIT_TO_STAND','SIT_TO_LIE','LIE_TO_SIT','STAND_TO_LIE','LIE_TO_STAND']
get_best_randomsearch_results(lr_classifier_rs)
# labels, title and ticks
```
